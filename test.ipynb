```python
# import  libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from IPython.display import Image,display
from tensorflow.keras.models import load_model
```

```python
# model path
model_path='/content/drive/MyDrive/model.h5'

# image path
image_path1='/content/drive/MyDrive/31fcdc161.png'
image_path2='/content/drive/MyDrive/43ede9de9.png'
image_path3='/content/drive/MyDrive/f8318faf1.png'

# define size of image
image_size=(224,224)
```

```python
# define class label names

class_label=['Black-grass',
 'Charlock',
 'Cleavers',
 'Common Chickweed',
 'Common wheat',
 'Fat Hen',
 'Loose Silky-bent',
 'Maize',
 'Scentless Mayweed',
 'Shepherds Purse',
 'Small-flowered Cranesbill',
 'Sugar beet']
```

```python
# function to get model

def get_model(path):
  return load_model(path)
```

```python
# function to get image

def get_image(img_path,img_size):
  img=keras.preprocessing.image.load_img(img_path,target_size=img_size)
  array=keras.preprocessing.image.img_to_array(img)
  array=np.expand_dims(array,axis=0)
  array=array/255.0
  return array
```

```python
# function to test new data in model

def test():
  model=get_model(model_path)
  image=get_image(image_path1,image_size)
  pred=model.predict(image)
  pred_index=np.argmax(pred,axis=1)
  print('the predicted class of image is : {}'.format(class_label[int(pred_index)]))
  print('\n')
  display(Image(image_path1))
```

```python
test()
```
